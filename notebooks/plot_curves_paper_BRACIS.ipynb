```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# data
data = {
    'Model': ['BERTimbau', 'BERTimbau', 'BERTimbau', 'BERTimbau', 'BERTimbau', 'DistilBERT', 'DistilBERT', 'DistilBERT', 'DistilBERT', 'DistilBERT', 'Pierreguillou', 'Pierreguillou', 'Pierreguillou', 'Pierreguillou', 'Pierreguillou', 'Linear Model',  'Linear Model',  'Linear Model',  'Linear Model',  'Linear Model',  'ChatGPT-3.5 \nPrompt#1', 'ChatGPT-3.5 \nPrompt#1', 'ChatGPT-3.5 \nPrompt#1', 'ChatGPT-3.5 \nPrompt#1', 'ChatGPT-3.5 \nPrompt#1', 'ChatGPT-3.5 \nPrompt#2','ChatGPT-3.5 \nPrompt#2','ChatGPT-3.5 \nPrompt#2','ChatGPT-3.5 \nPrompt#2','ChatGPT-3.5 \nPrompt#2'],
    'F1-Score': [0.7626, 0.7635, 0.7542, 0.7579, 0.7579, 0.7480, 0.7488, 0.7419, 0.7490, 0.7490, 0.7529, 0.7465, 0.7565, 0.7322, 0.7322, 0.5928, 0.6728, 0.6828, 0.6728, 0.6328, 0.71, 0.71, 0.71, 00.71, 0.71, 0.72, 0.72, 0.72, 0.72, 0.72],
    'Precision': [0.7308, 0.7354, 0.7160, 0.7172, 0.7172, 0.7361, 0.7237, 0.7152, 0.7241, 0.7241, 0.7243, 0.7321, 0.7148, 0.7382, 0.7382, 0.6744, 0.6744, 0.6644, 0.6844, 0.5944, 0.60, 0.60, 0.60, 0.60, 0.60, 0.78, 0.78, 0.78, 0.78, 0.78],
    'Recall': [0.8057, 0.7839, 0.8193, 0.8035, 0.8035, 0.7528, 0.7757, 0.7708, 0.7757, 0.7757, 0.7839, 0.7616, 0.8035, 0.7262, 0.7262, 0.6365, 0.6765, 0.7065, 0.6465, 0.6765, 0.87, 0.87, 0.87, 0.87, 0.87, 0.66, 0.66, 0.66, 0.66, 0.66]
}

# create DataFrame
df = pd.DataFrame(data)

# calculate means and standard deviations
means = df.groupby('Model').mean().reset_index()
# reset_index
stds = df.groupby('Model').std().reset_index()

# melt the dataframes to long format
melted_means = pd.melt(means, id_vars="Model", var_name="Metric", value_name="Mean")
melted_stds = pd.melt(stds, id_vars="Model", var_name="Metric", value_name="Std")

# merge the means and stds dataframes
df_final = pd.merge(melted_means, melted_stds, on=["Model", "Metric"])

# set color palette to grayscale
palette = sns.color_palette("gray", 3)

# create barplot
plt.figure(figsize=(12, 6))
sns.barplot(x="Model", y="Mean", hue="Metric", data=df_final, palette=palette, capsize=0.1)

# plot error bars
models = df_final["Model"].unique()
for i, model in enumerate(models):
    model_data = df_final[df_final["Model"] == model]
    for j, metric in enumerate(model_data["Metric"].unique()):
        mean = model_data[model_data["Metric"] == metric]["Mean"]
        std = model_data[model_data["Metric"] == metric]["Std"]
        plt.errorbar([i + (j-1)*0.25], mean, yerr=std, fmt='none', c='black', capsize=5)

# set labels
plt.xlabel('Models', fontweight='bold', fontsize=15)
plt.ylabel('Scores', fontweight='bold', fontsize=15)
plt.legend(title='Metric')
plt.title('Model Performance', fontweight='bold', fontsize=15)

plt.show()

```
