```python
import pandas as pd

# Carregar a base de dados
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/2023/BRACIS_toxic/data/DOX_RAW_data_test_xenofobia_annot.csv')

# Verificar o consenso entre as colunas 'anotadora1' e 'anotadora2'
df['consenso'] = df.apply(lambda row: 1 if row['anotadora1'] == row['anotadora2'] else None, axis=1)

# Remover as linhas onde não houve consenso
df = df.dropna(subset=['consenso'])

# Mapear os rótulos 'nao-xenofobia' para 0 e 'xenofobia' para 1
df['target'] = df['anotadora1'].map({'nao-xenofobia': 0, 'xenofobia': 1})

# Gerar a matriz de variável independente (X) e o vetor target (Y)
X = df['clean_text']
Y = df['target']

# Criar um novo DataFrame
new_df = pd.DataFrame({
    'clean_text': X,
    'target': Y
})

# Salvar o novo DataFrame em um arquivo CSV
new_df.to_csv('/content/drive/MyDrive/Colab Notebooks/2023/BRACIS_toxic/data/DOX_teste_processado.csv', index=False)

```
